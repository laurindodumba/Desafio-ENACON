```python
import pandas as pd
import glob


# Carregamentos da base de dados, fiz a concatenao
arquivos_csv = glob.glob('/content/dados/GERACAO_USINA_*.csv')

# Lista para armazenar DataFrames individuais
dfs = []

# Itera sobre cada arquivo CSV, l-o como DataFrame e adiciona  lista dfs
for arquivo in arquivos_csv:
    df = pd.read_csv(arquivo, sep=';')  # Especifica o delimitador como ';'
    dfs.append(df)

# Concatena todos os DataFrames da lista dfs em um nico DataFrame
df_final = pd.concat(dfs, ignore_index=True)

# Verifica o resultado
print(df_final.head())
```

```python
def cons_temp(df_final):
    df_final['din_instante'] = pd.to_datetime(df_final['din_instante'], format='%Y-%m-%d %H:%M:%S')

    df_final['Mes'] = df_final['din_instante'].dt.month  # Corrigido para dt.month
    df_final['Semana'] = df_final['din_instante'].dt.weekday  # Corrigido para dt.weekday
    df_final['Dia'] = df_final['din_instante'].dt.day
    df_final['Hora'] = df_final['din_instante'].dt.hour

    return df_final
```

```python
df_final = cons_temp(df_final)
```

# 03 - REALIZAO A PREDIO USANDO O ALGORITMO PROPHET

```python
df_pred = df_final.rename(columns={'din_instante': 'ds',
                        'val_geracao': 'y'})

df_pred.head()
```

```python
df_pred = df_pred[['y', 'ds']]
```

```python
df_pred.head()
```

```python
# Python
import pandas as pd
from prophet import Prophet

```

```python
# Converter  a coluna da data para 'ds' do DataFrame Original
df_pred['ds'] = pd.to_datetime(df_pred['ds'])

df_filtro = df_pred[(df_pred['ds'].dt.year >= 2018) & (df_pred['ds'].dt.year <= 2019)]

df_pred = df_pred[['y', 'ds']]

#Realizao da predio
m = Prophet()
m.fit(df_filtro)
```

```python
# Realizando a predio os prximos 12 meses
future = m.make_future_dataframe(periods=12)
future.tail()
```

```python
# Python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Python
fig1 = m.plot(forecast)

```

```python
# Python
fig2 = m.plot_components(forecast)
```

```python
# Python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)
```
